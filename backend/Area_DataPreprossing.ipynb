```python
import pandas as pd
df = pd.read_csv("area_addiction_dataset.csv")
```

```python
state_cities = (
    df.groupby("State")["City"]
      .unique()               # unique city list
      .apply(lambda x: x[:5]) # first 5 cities
)

# Display the result
for state, cities in state_cities.items():
    print(f"State: {state}")
    for city in cities:
        print(f"  - {city}")
    print()
```
